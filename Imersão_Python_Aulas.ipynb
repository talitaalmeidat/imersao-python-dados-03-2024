```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/acoes_pura.xlsx",sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/acoes_pura.xlsx",sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/acoes_pura.xlsx",sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/acoes_pura.xlsx",sheet_name="Chatgpt")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)','Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)': 'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal ['Var_pct'] = df_principal['var_dia_pct']/100
df_principal['valor_inicial'] = df_principal['valor_final']/ (df_principal ['Var_pct']+1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes,left_on='Ativo',right_on='Código',how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])* df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format='{:.2f}'.format
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica': 'qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker,left_on='Ativo',right_on='Ticker',how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt,left_on='Nome',right_on='Empresa',how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
df_principal['Cat_Idade'] = df_principal['Idade (anos)'].apply(lambda x: 'Mais que 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python


# Calcular a média
media = df_principal['Variacao_rs'].mean()

# Calcular a média de quem subiu
# == comparação = atribuição

media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calcular a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Calcular o maior valor
maior_valor = df_principal['Variacao_rs'].max()

# Calcular o menor valor
menor_valor = df_principal['Variacao_rs'].min()

print("Maior: R$", maior_valor)
print("Menor: R$", menor_valor)
print("Média: R$", media)
print("Média de quem subiu: R$", media_subiu)
print("Média de quem desceu: R$", media_desceu)
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
#Agrupamento por segmento
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
#Saldo
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python

# Criando o gráfico
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Resultado')

# Atualizando as informações de texto para formatá-las
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

# Exibindo o gráfico
fig.show()
```

```python
#Fazer o gráfico de pizza no df_análise_segmentos com a mesma biblioteca Potly;
# Criar o gráfico de pizza
fig = px.pie(df_analise_segmento, values='Variacao_rs', names='Segmento', title='Variação Reais por Segmento')

# Exibir o gráfico
fig.show()
```

```python
#Fazer o GroupBy da categoria de idades e gerar o gráfico de barras.
df_analise_cat_idade = df_principal.groupby('Cat_Idade')['Variacao_rs'].sum().reset_index()
df_analise_cat_idade['Qtd_empresas']= ''
df_analise_cat_idade
```

```python
df_analise_idade = df_principal.groupby('Cat_Idade')['Variacao_rs'].sum().reset_index()
df_analise_idade

df_qte_empresas = df_principal.groupby('Cat_Idade').size().reset_index()
df_qte_empresas = df_qte_empresas.rename(columns = {0 : 'Qtd_empresas'}).copy()
df_qte_empresas
```

```python
fig = px.bar(df_analise_cat_idade, x='Cat_Idade', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Categoria de Idade')
fig.show()
```

```python
fig = px.bar(df_qte_empresas, x='Cat_Idade', y='Qtd_empresas', text='Qtd_empresas', title='Análise por faixa etária')
fig.show()
```
